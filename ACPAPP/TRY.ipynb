```python
# %%
# Import of the pyomo module
from pyomo.environ import *
from pyomo.opt import SolverFactory
import pyomo.environ

# Creation of a Concrete Model
model = ConcreteModel()

# %%
model.K = Set(initialize=range(m), doc='interval de lindice k')

model.I = Set(initialize=range(0,n), doc='interval de lindice i et J')
model.J = Set(initialize=range(1,n), doc='Quantity conducteur')
# %%

model.c = Param(model.K,initialize= c, doc='capacite de chaque véhicule')
model.q = Param(model.I,initialize= q, doc='quantite du conteneur i')
model.d = Param(model.I,model.I,initialize=d, domain= Any ,doc=' distance entre le point i et j')
# %%
model.x = Var(model.I, model.I,model.K,domain=Binary, doc='variable de decision boolean prend 1 si le vehicule 0 parcourt l arrete (i,j)')
model.y = Var(model.I,model.K,bounds=(0,None),domain=Integers, doc='ordre de passage')

# %%
def Cotraint1(model,j):
  return sum(model.x[i,j,k] for i in [l for l in model.I if l != j] for k in model.K)==1
model.C1 = Constraint(model.J, rule=Cotraint1, doc='Chaque vehicule si utiliser demarre du depot')


def Contraint2(model,k):
  return sum(model.x[0,j,k] for j in model.J)==1

model.C2 = Constraint ( model.K, rule=Contraint2, doc='verification de non chevauchement ')

def Contraint3(model,j,k):
  return sum(model.x[i,j,k] for i in [l for l in model.I if l != j] )==sum(model.x[j,i,k] for i in [l for l in model.I if l != j])

model.C3 = Constraint (model.J,model.K, rule=Contraint3, doc='  Chaque point visite doit etre quitter.')

def Contraint4(model,k):
  return  sum((model.x[i,j,k]*q[j]) for i in model.I for j in [l for l in model.J if l != i])<=model.c[k]

model.C4 = Constraint (model.K, rule=Contraint4, doc='La capacite de chaque vehicule doit etre respecter.')

def Contraint5(model,i,j ,k):
  return (model.y[i,k]+1)<=(model.y[j,k]+n*(1-model.x[i,j,k])) 
model.C5 = Constraint(model.I,model.J,model.K, rule=Contraint5, doc='verification daffectation de terminal')

# %%

# %%
def objective_rule(model):
  return sum(model.x[i,j,k]*model.d[i,j] for i in model.I for j in [l for l in model.I if l != i] for k in model.K)
model.objective = Objective(rule=objective_rule, sense=minimize, doc='Define objective function')

# %%
def pyomo_postprocess(options=None, instance=None, results=None):
  model.x.display()
# %%


opt = SolverFactory("gurobi")
results = opt.solve(model,tee=True)
#sends results to stdout
results.write()
print("\nDisplaying Solution\n" + '-'*60)
pyomo_postprocess(None, model, results)
# %%
```
